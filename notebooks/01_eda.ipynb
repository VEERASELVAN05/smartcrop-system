```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(r"C:\Users\mages\capstone\smartcrop-system\data\raw\crop_recommendation.csv")

# Basic exploration
print("Shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
print("\nColumn names:", df.columns.tolist())
print("\nData types:")
print(df.dtypes)
print("\nMissing values:")
print(df.isnull().sum())
print("\nBasic statistics:")
print(df.describe())
```

```python
plt.figure(figsize=(14, 6))
df['label'].value_counts().plot(kind='bar', color='#27ae60')
plt.title('Crop Distribution in Dataset', fontsize=16)
plt.xlabel('Crop Type', fontsize=12)
plt.ylabel('Number of Samples', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(r"C:\Users\mages\capstone\smartcrop-system\docs\crop_distribution.png")
plt.show()
print("Saved!")
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['rainfall'], bins=30, color='#2196F3')
plt.title('Rainfall Distribution Across Crops', fontsize=16)
plt.xlabel('Rainfall (mm)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()
plt.savefig(r"C:\Users\mages\capstone\smartcrop-system\docs\rainfall_distribution.png")
plt.show()
print("Saved!")
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.drop('label', axis=1).corr(), 
            annot=True, 
            cmap='coolwarm', 
            fmt='.2f',
            linewidths=0.5)
plt.title('Feature Correlation Heatmap', fontsize=16)
plt.tight_layout()
plt.savefig(r"C:\Users\mages\capstone\smartcrop-system\docs\correlation_heatmap.png")
plt.show()
print("Saved!")
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='temperature', 
                y='rainfall', 
                hue='label', 
                data=df,
                alpha=0.7)
plt.title('Temperature vs Rainfall by Crop Type', fontsize=16)
plt.xlabel('Temperature (°C)', fontsize=12)
plt.ylabel('Rainfall (mm)', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig(r"C:\Users\mages\capstone\smartcrop-system\docs\temp_vs_rainfall.png")
plt.show()
print("Saved!")
```

## EDA Key Observations

1. Dataset has 2200 samples across 22 different crops
2. Each crop has exactly 100 samples — balanced dataset
3. Rainfall ranges from 20mm to 298mm across different crops
4. Temperature ranges from 8°C to 43°C
5. Rice requires highest rainfall (avg ~236mm)
6. No missing values found — clean dataset
7. Strong correlation between humidity and rainfall
8. NPK values vary significantly between crop types

## What This Means for Our Model
- Rainfall and temperature are the strongest predictors
- Soil NPK values help differentiate crop types
- Dataset is well balanced — no need for oversampling
